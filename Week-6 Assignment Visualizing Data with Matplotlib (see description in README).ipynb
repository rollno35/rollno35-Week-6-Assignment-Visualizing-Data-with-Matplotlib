```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
data.rename(columns={'Study Hours': 'Study_Hours', 'Socioeconomic Score': 'Socioeconomic_Score', 'Sleep Hours': 'Sleep_Hours'}, inplace=True)
```

```python
data.head()
```

<font size="5">Line Plot</font>

```python
plt.plot(data['Study_Hours'], data['Grades'], linewidth='0.1', color='hotpink', marker='.', ms='1', mec='palevioletred', linestyle='-')
#'Study_Hours' in x
#'Grades' in y
#we have a lot of points and lines here so the smalller and the thinner the better
#using linewidth to make the lines thinner
#color for line color
# . as marker
#ms for marker size
#mec colors marker edges
#linestyle - 
plt.title('Hours of Study vs Grades')
plt.xlabel("Hours studied per day")
plt.ylabel('Grades obtained')
plt.grid(True)
plt.show()
```

 Insights and Observations:
No studied got grade less than 30. The students who study less than two hours do not have good grades, their grades are below 40. The students who study between two to four hours typically have grades under 50, with only a few students who studying less than four hours achieving grades above 50. The students who study between four to six hours shows result between 30ish to even more than 60 so the range is quite wide here. The students who study more than six hours a day tend to get good grades. The students who study more than eight hours mostly achieve even better grades, ranging from 50 to 80. The student with the highest grades which is above 90, studies almost 10 hours a day. However, The student who studied exactly 10 hours a day got marks around 65. 
Most students have grades in the 30 to 50 range, and only a few students achieve very high scores, with most of the high score ones studies more than eight hours a day.

So the more time someone spends on studying the more likely are they to get higher grades. But there are a lot of exceptions of this rule.


<font size="5">Bar Chart</font>

```python
plt.bar(data['Socioeconomic_Score'],data['Grades'],width=0.1,color='mistyrose',edgecolor='black')
#x axis column , y axis column, width, color, edgecolor
plt.title('Socioeconomic Score vs Grades')
plt.xlabel("Socioeconomic Score")
plt.ylabel("Grades")
plt.show()
```

 Insights and Observations:
It can be noticed that up to a socioeconomic score of 0.8, there didn't really seem to be any difference in the grades, but when the socioeconomic score became more than 0.8, which is really high because the maximum is 1, the grades seemed to get higher, and the students with the highest grades have a socioeconomic state of 1.


<font size="5">Scatter Plot</font>

```python
colors = np.random.randint(100, size=(1388))
plt.scatter(data['Sleep_Hours:'],data['Grades'], c=colors, cmap="pink" , marker="*", alpha=0.5) #using colormap from matplotlib buiilt in colormaps, alpha for transparency
plt.title('Hours of Sleep vs Grades')
plt.xlabel("Hours of Sleep")
plt.ylabel("Grades")
plt.show()
```

Insights and Observations:
The hours of sleep do not have any sort of impact on the grades.

```python
colors = np.random.randint(100, size=(1388))
plt.scatter(data['Attendance '],data['Grades'], c=colors, cmap="Pastel1" , marker="*", alpha=0.9) #using colormap from matplotlib buiilt in colormaps, alpha for transparency
plt.title('Attendance vs Grades')
plt.xlabel("Attendance")
plt.ylabel("Grades")
plt.show()
```

Insights and Observations:
It's actually interesting that attendance has almost no effect on grade. The students with very high attendance performed almost similarly to those with low attendance. And the students who scored highest(90 ish) have low attendance between 40 to 60.

<font size="5">Box PLot</font>

```python
plt.title('BoxPlot of Grades')
plt.boxplot(data['Grades'],patch_artist=True,boxprops=dict(facecolor='pink',color="black"),whiskerprops=dict(color="pink"),capprops=dict(color="skyblue",linewidth=1),flierprops=dict(marker='*', color="skyblue", markersize=5))
#patch_artist to fill the box with colors
#...props for styling
plt.ylabel("Grades")
plt.grid(True)
plt.show()
```

Insights and Observations:
The student with the lowest score had little more than 30, and most of the students scored between 35 to a little more than 45. However, there were a few student who scored above 70, even up to 90. But it's still very rare when compared  to the number of students who scored somewhere between 30 to 50.
